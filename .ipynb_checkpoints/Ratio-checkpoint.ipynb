```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("malicious_url.csv")
```

```python
dataset['Uppercase'] = dataset['url'].str.count(r'[A-Z]')
dataset['Lowercase'] = dataset['url'].str.count(r'[a-z]')
```

```python
dataset.head()
```

```python
dataset['Ratio of Ucase to Lcase'] = dataset['Uppercase'] / dataset['Lowercase']
```

```python
dataset.head()
```

```python
dataset['Alpha_len'] = (dataset['url'].str.count(r'[A-Z]'))+(dataset['url'].str.count(r'[a-z]'))
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset['count_numbers'] = dataset['url'].str.count('\d')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
def D_C_ratio(url):
    count_numbers= sum(item.isdigit() for item in url)
    Char_len=len(url)
    r = count_numbers/Char_len
    return(r)
```

```python
dataset['Digits_to_char_ratio'] = dataset['url'].apply(D_C_ratio)
```

```python
dataset.head()
```

```python
def U_C_ratio(url):
    Uppercase= sum(item.isupper() for item in url)
    length=len(url)
    ratio = Uppercase/length
    return(ratio)
```

```python
dataset['Ucase_to_char_ratio'] = dataset['url'].apply(U_C_ratio)
```

```python
dataset.head()
```

```python
def L_C_ratio(url):
    Lowercase= sum(item.islower() for item in url)
    length=len(url)
    ratio = Lowercase/length
    return(ratio)
```

```python
dataset['Lcase_to_char_ratio'] = dataset['url'].apply(L_C_ratio)
```
